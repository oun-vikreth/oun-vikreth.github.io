::: {.callout-tip title="Problem 2.4" appearance="simple"}

Let $\{e_t\}$ be a zero mean white noise process. Suppose that the observed process is $Y_t = e_t + \theta e_{t-1}$, where $\theta$ is either $3$ or $\frac{1}{3}$.

a) Find the autocorrelation function for $\{Y_t\}$ both when $\theta = 3$ and when $\theta = \frac{1}{3}$.

b) You should have discovered that the time series is stationary regardless of the value of $\theta$ and that the autocorrelation functions are the same for $\theta = 3$ and $\theta = \frac{1}{3}$. For simplicity, suppose that the process mean is known to be zero and the variance of $Y_t$ is known to be $1$. You observe the series $\{Y_t\}$ for $t = 1, 2, \ldots, n$ and suppose that you could determine which value of $\theta$ is correct ($3$ or $\frac{1}{3}$) based on the estimate of $\rho_k$? Why and why not?

:::

### **Solution 2.4**

Let $e_t$ be a zero-mean white noise process. Suppose the observed process is
$$
Y_t = e_t + \theta e_{t-1},
$$
where $\theta$ is either $3$ or $\frac{1}{3}$.

### (a) Find the autocorrelation function for $Y_t$ when $\theta = 3$ and $\theta = \frac{1}{3}$.

The autocovariance function $\gamma(k)$ for $Y_t$ is defined as:
$$
\gamma(k) = \mathrm{Cov}(Y_t, Y_{t-k}).
$$

**For $k = 0$ (variance):**
$$
\gamma(0) = \mathrm{Var}(Y_t) = \mathrm{Var}(e_t) + \theta^2 \mathrm{Var}(e_{t-1}),
$$
where $\mathrm{Var}(e_t) = \sigma^2$. Thus:
$$
\gamma(0) = \sigma^2 (1 + \theta^2).
$$

**For $k = 1$ (lag-1 autocovariance):**
$$
\gamma(1) = \mathrm{Cov}(Y_t, Y_{t-1}) = \mathrm{Cov}(e_t + \theta e_{t-1}, e_{t-1} + \theta e_{t-2}).
$$
Using the independence of $e_t$, $e_{t-1}$, and $e_{t-2}$:
$$
\gamma(1) = \mathrm{Cov}(e_t, e_{t-1}) + \theta \mathrm{Cov}(e_t, e_{t-2}) + \theta \mathrm{Cov}(e_{t-1}, e_{t-1}) + \theta^2 \mathrm{Cov}(e_{t-1}, e_{t-2}).
$$
Since $e_t$ is white noise, only $\mathrm{Cov}(e_{t-1}, e_{t-1}) = \mathrm{Var}(e_{t-1}) = \sigma^2$ contributes:
$$
\gamma(1) = \theta \sigma^2.
$$

**For $k \geq 2$:**
$$
\gamma(k) = 0,
$$
because $e_t$ is white noise, and there are no overlapping terms.

:::: {.column-margin}
::: {.callout-note}

A time series model $Y_t$ which is weakly stationary with

$\mathbb{E}[e_t] = 0$

$\text{Cov}(e_t, e_{t+k}) = \begin{cases}
\sigma^2, & \text{if } k=0 \\
0, & \text{if } k \neq 0
\end{cases}$

is said to be white noise with variance $\sigma^2$.

:::
::::

The autocorrelation function $\rho(k)$ is given by:
$$
\rho(k) = \frac{\gamma(k)}{\gamma(0)}.
$$

**For $k = 0$:**
$$
\rho(0) = 1.
$$

**For $k = 1$:**
$$
\rho(1) = \frac{\gamma(1)}{\gamma(0)} = \frac{\theta \sigma^2}{\sigma^2 (1 + \theta^2)} = \frac{\theta}{1 + \theta^2}.
$$

**For $k \geq 2$:**
$$
\rho(k) = 0.
$$

**Results:**

- When $\theta = 3$:
$$
\rho(1) = \frac{3}{1 + 3^2} = \frac{3}{10}.
$$

- When $\theta = \frac{1}{3}$:
$$
\rho(1) = \frac{\frac{1}{3}}{1 + \left(\frac{1}{3}\right)^2} = \frac{\frac{1}{3}}{1 + \frac{1}{9}} = \frac{\frac{1}{3}}{\frac{10}{9}} = \frac{3}{10}.
$$

Thus, the autocorrelation function is the same for both $\theta = 3$ and $\theta = \frac{1}{3}$.

### (b) Can we determine the correct value of $\theta$ based on the estimate of $\rho_k$?

No, it is not possible to determine which value of $\theta$ is correct based solely on the estimate of $\rho_k$. This is because the autocorrelation function $\rho(k)$ is the same for both $\theta = 3$ and $\theta = \frac{1}{3}$. 

::: {.callout-note appearance="simple"}

## Pay Attention

The symmetry in the formula for $\rho(k)$ arises due to the quadratic term $1 + \theta^2$ in the denominator, which makes $\rho(k)$ identical for both values of $\theta$. Thus, the observed series $\{Y_t\}$ will exhibit identical second-order properties (variance and autocorrelation) for both values of $\theta$, making them indistinguishable based on $\rho_k$ alone.

**Conclusion:** Additional information or a different approach (e.g., higher-order moments, maximum likelihood estimation, or model selection criteria) would be required to distinguish between $\theta = 3$ and $\theta = \frac{1}{3}$.

:::

::: {.callout-tip title="Problem 2.5" appearance="simple"}

Suppose $Y_t = 5+2t+X_t$, where ${X_t}$ is a zero-mean stationary series with auto
covariance function $\gamma_k$.

 a) Find the mean function for ${Y_t}$.

 b) Find the autocovariance function for ${Y_t}$.
 
 c) Is ${Y_t}$ stationary? Why or why not? 

:::

### **Solution 2.5**

Suppose $Y_t = 5 + 2t + X_t$, where $\{X_t\}$ is a zero-mean stationary series with autocovariance function $\gamma_k$.

### (a) Find the mean function for $\{Y_t\}$.

The mean function $\mu_Y(t)$ for $\{Y_t\}$ is defined as:
$$
\mu_Y(t) = \mathbb{E}[Y_t].
$$

Substitute $Y_t = 5 + 2t + X_t$:
$$
\mu_Y(t) = \mathbb{E}[5 + 2t + X_t].
$$

Using the linearity of expectation and the fact that $\mathbb{E}[X_t] = 0$ (since $\{X_t\}$ is zero-mean):
$$
\mu_Y(t) = 5 + 2t + \mathbb{E}[X_t] = 5 + 2t.
$$

Thus, the mean function is:

$$
\mu_Y(t) = 5 + 2t.
$$

### (b) Find the autocovariance function for $\{Y_t\}$.

The autocovariance function $\gamma_Y(k)$ for $\{Y_t\}$ is defined as:
$$
\gamma_Y(k) = \mathrm{Cov}(Y_t, Y_{t-k}).
$$

Substitute $Y_t = 5 + 2t + X_t$:
$$
\gamma_Y(k) = \mathrm{Cov}(5 + 2t + X_t, 5 + 2(t-k) + X_{t-k}).
$$

Using the linearity of covariance and the fact that constants do not contribute to covariance:
$$
\gamma_Y(k) = \mathrm{Cov}(X_t, X_{t-k}).
$$

Since $\{X_t\}$ is stationary with autocovariance function $\gamma_k$:
$$
\gamma_Y(k) = \gamma_k.
$$

Thus, the autocovariance function of $\{Y_t\}$ is the same as that of $\{X_t\}$.

### (c) Is $\{Y_t\}$ stationary? Why or why not?

:::: {.column-margin}
::: {.callout-note}

The Time series are stationarity if:
- its mean and variance are constant over time
- covariance depends only on the distance between the two periods and not on time
- a time series with these characteristics is know as weakly or covariancestationary

:::
::::

- The mean function of $\{Y_t\}$ is $\mu_Y(t) = 5 + 2t$, which depends on $t$. Since the mean changes with time, $\{Y_t\}$ is \textbf{not stationary}.
- The variance and autocovariance functions of $\{Y_t\}$ are the same as those of $\{X_t\}$, which are stationary. However, stationarity requires all statistical properties, including the mean, to be time-invariant.

Thus, $\{Y_t\}$ is **not stationary** because its mean function is time-dependent.

::: {.callout-tip title="Problem 2.8" appearance="simple"}

2.8 Suppose that $\{Y_t\}$ is stationary with autocovariance function $\gamma_k$. Show that for any fixed positive integer $n$ and any constants $c_1, c_2, ..., c_n$, the process $\{W_t\}$ defined by $$W_t = c_1Y_t + c_2Y_{t-1} + ... + c_nY_{t-n+1}$$ is stationary.

:::

### **Solution 2.8**

### (a) Mean Function

The mean function of $\{W_t\}$ is:
$$
\mathbb{E}[W_t] = \mathbb{E}[c_1 Y_t + c_2 Y_{t-1} + \cdots + c_n Y_{t-n+1}].
$$

Using the linearity of expectation:
$$
\mathbb{E}[W_t] = c_1 \mathbb{E}[Y_t] + c_2 \mathbb{E}[Y_{t-1}] + \cdots + c_n \mathbb{E}[Y_{t-n+1}].
$$

Since $\{Y_t\}$ is stationary, $\mathbb{E}[Y_t]$ is constant (say $\mu$). Thus:
$$
\mathbb{E}[W_t] = c_1 \mu + c_2 \mu + \cdots + c_n \mu = \mu (c_1 + c_2 + \cdots + c_n).
$$

Since the result is independent of $t$, the mean function of $\{W_t\}$ is constant.

### (b) Variance

The variance of $\{W_t\}$ is:
$$
\mathrm{Var}(W_t) = \mathbb{E}[W_t^2] - (\mathbb{E}[W_t])^2.
$$

Substituting $W_t = c_1 Y_t + c_2 Y_{t-1} + \cdots + c_n Y_{t-n+1}$:
$$
\mathbb{E}[W_t^2] = \mathbb{E} \left[ \left( c_1 Y_t + c_2 Y_{t-1} + \cdots + c_n Y_{t-n+1} \right)^2 \right].
$$

Expanding the square:
$$
\mathbb{E}[W_t^2] = \sum_{i=1}^n \sum_{j=1}^n c_i c_j \mathbb{E}[Y_{t-i+1} Y_{t-j+1}].
$$

The expectation $\mathbb{E}[Y_{t-i+1} Y_{t-j+1}]$ depends only on the lag $|i - j|$ because $\{Y_t\}$ is stationary. Let $\gamma_{|i-j|}$ denote the autocovariance at lag $|i-j|$. Then:
$$
\mathbb{E}[W_t^2] = \sum_{i=1}^n \sum_{j=1}^n c_i c_j \gamma_{|i-j|}.
$$

The variance $\mathrm{Var}(W_t)$ is independent of $t$ because $\gamma_{|i-j|}$ depends only on the lag $|i-j|$.

### (c) Autocovariance Function

The autocovariance of $\{W_t\}$ is:
$$
\mathrm{Cov}(W_t, W_{t-k}) = \mathbb{E}[W_t W_{t-k}] - \mathbb{E}[W_t] \mathbb{E}[W_{t-k}].
$$

Substituting $W_t = c_1 Y_t + c_2 Y_{t-1} + \cdots + c_n Y_{t-n+1}$ and $W_{t-k} = c_1 Y_{t-k} + c_2 Y_{t-k-1} + \cdots + c_n Y_{t-k-n+1}$:
$$
\mathbb{E}[W_t W_{t-k}] = \mathbb{E} \left[ \left( \sum_{i=1}^n c_i Y_{t-i+1} \right) \left( \sum_{j=1}^n c_j Y_{t-k-j+1} \right) \right].
$$

Expanding the product:
$$
\mathbb{E}[W_t W_{t-k}] = \sum_{i=1}^n \sum_{j=1}^n c_i c_j \mathbb{E}[Y_{t-i+1} Y_{t-k-j+1}].
$$

The expectation $\mathbb{E}[Y_{t-i+1} Y_{t-k-j+1}]$ depends only on the lag $|(i-j) + k|$ because $\{Y_t\}$ is stationary. Let $\gamma_{|(i-j)+k|}$ denote the autocovariance at lag $|(i-j)+k|$. Then:
$$
\mathrm{Cov}(W_t, W_{t-k}) = \sum_{i=1}^n \sum_{j=1}^n c_i c_j \gamma_{|(i-j)+k|}.
$$

Since $\gamma_{|(i-j)+k|}$ depends only on the lag $k$ and not on $t$, the autocovariance of $\{W_t\}$ depends only on the lag $k$.

The process $\{W_t\}$ has:
1. A constant mean function.
2. A constant variance.
3. An autocovariance function that depends only on the lag.

Thus, $\{W_t\}$ is stationary.

::: {.callout-tip title="Problem 2.9" appearance="simple"}

Suppose $Y_t = \beta_0 + \beta_1 t + X_t$, where $\{X_t\}$ is a zero-mean stationary series with autocovariance function $\gamma_k$ and $\beta_0$ and $\beta_1$ are constants.

a) Show that $\{Y_t\}$ is not stationary but $W_t = \nabla Y_t = Y_t - Y_{t-1}$ is stationary.

b) In general, show that if $Y_t = \mu_t + X_t$, where ${X_t}$ is a zero-mean stationary series and $\mu_t$ is a polynomial in $t$ of degree $d$, then $\nabla^m Y_t = \nabla(\nabla^{m−1}Y_t)$ is stationary for $m \geq d$ and nonstationary for $0  \leq m < d$.

:::

### Solution 2.9

### (a) Show that $\{Y_t\}$ is not stationary but $W_t = \nabla Y_t = Y_t - Y_{t-1}$ is stationary.

The process $Y_t$ is given by:
$$
Y_t = \beta_0 + \beta_1 t + X_t.
$$

#### Mean Function of $\{Y_t\}$:
The mean of $Y_t$ is:
$$
\mathbb{E}[Y_t] = \mathbb{E}[\beta_0 + \beta_1 t + X_t].
$$
Since $\{X_t\}$ is a zero-mean stationary process, $\mathbb{E}[X_t] = 0$. Thus:
$$
\mathbb{E}[Y_t] = \beta_0 + \beta_1 t.
$$

The mean function depends on $t$, so $\{Y_t\}$ is not stationary.

#### Differenced Process $\{W_t\}$:
The differenced process is:
$$
W_t = \nabla Y_t = Y_t - Y_{t-1}.
$$

Substituting $Y_t = \beta_0 + \beta_1 t + X_t$:
$$
W_t = (\beta_0 + \beta_1 t + X_t) - (\beta_0 + \beta_1 (t-1) + X_{t-1}).
$$
Simplify:
$$
W_t = \beta_1 + (X_t - X_{t-1}).
$$

#### Mean Function of $\{W_t\}$:
The mean of $W_t$ is:
$$
\mathbb{E}[W_t] = \mathbb{E}[\beta_1 + (X_t - X_{t-1})].
$$
Since $\mathbb{E}[X_t] = \mathbb{E}[X_{t-1}] = 0$:
$$
\mathbb{E}[W_t] = \beta_1.
$$
The mean is constant, independent of $t$.

#### Autocovariance Function of $\{W_t\}$:
The autocovariance at lag $k$ is:
$$
\mathrm{Cov}(W_t, W_{t-k}) = \mathrm{Cov}(\beta_1 + (X_t - X_{t-1}), \beta_1 + (X_{t-k} - X_{t-k-1})).
$$
Since $\beta_1$ is constant:
$$
\mathrm{Cov}(W_t, W_{t-k}) = \mathrm{Cov}(X_t - X_{t-1}, X_{t-k} - X_{t-k-1}).
$$
Expanding the covariance:
$$
\mathrm{Cov}(W_t, W_{t-k}) = \mathrm{Cov}(X_t, X_{t-k}) - \mathrm{Cov}(X_t, X_{t-k-1}) - \mathrm{Cov}(X_{t-1}, X_{t-k}) + \mathrm{Cov}(X_{t-1}, X_{t-k-1}).
$$
Since $\{X_t\}$ is stationary, the autocovariance depends only on the lag:
$$
\mathrm{Cov}(W_t, W_{t-k}) = \gamma_k - \gamma_{k+1} - \gamma_{k-1} + \gamma_k.
$$
Thus, $\{W_t\}$ has a constant autocovariance function, and $\{W_t\}$ is stationary.

### (b) General Case: $Y_t = \mu_t + X_t$, where $\mu_t$ is a polynomial of degree $d$.

Let $\mu_t$ be a polynomial in $t$ of degree $d$:
$$
\mu_t = a_0 + a_1 t + a_2 t^2 + \cdots + a_d t^d.
$$

#### Differencing Process:
The first difference is:
$$
\nabla Y_t = Y_t - Y_{t-1}.
$$
Substituting $Y_t = \mu_t + X_t$:
$$
\nabla Y_t = (\mu_t - \mu_{t-1}) + (X_t - X_{t-1}).
$$
The differencing reduces the degree of the polynomial $\mu_t$ by 1. After $m$ differences:
$$
\nabla^m Y_t = \nabla^m \mu_t + \nabla^m X_t.
$$

If $m \geq d$, $\nabla^m \mu_t$ becomes constant (or zero if $m > d$), leaving only $\nabla^m X_t$:
$$
\nabla^m Y_t = \nabla^m X_t.
$$
Since $\{X_t\}$ is stationary, $\nabla^m Y_t$ is stationary for $m \geq d$.

#### Nonstationarity for $m < d$:
If $m < d$, $\nabla^m \mu_t$ remains a polynomial in $t$ of degree $(d - m)$, which depends on $t$. Thus, $\nabla^m Y_t$ is nonstationary for $m < d$.

1. $\{Y_t\}$ is nonstationary because its mean depends on $t$.
2. $\nabla^m Y_t$ is stationary for $m \geq d$ and nonstationary for $0 \leq m < d$.

::: {.callout-tip title="Problem 2.12" appearance="simple"}

 Suppose that $Y_t = e_t - e_{t-12}$. Show that ${Y_t}$ is stationary and that, for $k > 0$, its
 autocorrelation function is nonzero only for lag $k = 12$.

:::

### Solution 2.12

Let's show that the process $Y_t = e_t - e_{t-12}$ is stationary and compute its autocorrelation function.

### Step 1: Stationarity of the Process

To show that the process $Y_t = e_t - e_{t-12}$ is stationary, we need to check if its mean, variance, and autocovariance are time-invariant.

#### Mean of $Y_t$

Since $e_t$ is a white noise process with zero mean, we have:

$$
\mathbb{E}[e_t] = 0 \quad \text{for all } t.
$$

Thus, the mean of $Y_t$ is:

$$
\mathbb{E}[Y_t] = \mathbb{E}[e_t - e_{t-12}] = \mathbb{E}[e_t] - \mathbb{E}[e_{t-12}] = 0 - 0 = 0.
$$

Since the mean of $Y_t$ is constant and does not depend on $t$, the process $Y_t$ has a constant mean, which is a necessary condition for stationarity.

#### Variance of $Y_t$

Next, we calculate the variance of $Y_t$. The variance is defined as:

$$
\text{Var}(Y_t) = \mathbb{E}[Y_t^2] - (\mathbb{E}[Y_t])^2.
$$

Since we have already shown that $\mathbb{E}[Y_t] = 0$, we just need to calculate $\mathbb{E}[Y_t^2]$:

$$
\mathbb{E}[Y_t^2] = \mathbb{E}[(e_t - e_{t-12})^2].
$$

Expanding the square:

$$
\mathbb{E}[Y_t^2] = \mathbb{E}[e_t^2 - 2e_t e_{t-12} + e_{t-12}^2].
$$

Since $e_t$ is a white noise process, we know that $\mathbb{E}[e_t e_{t-12}] = 0$ for $t \neq t-12$, so the cross-term vanishes. Thus:

$$
\mathbb{E}[Y_t^2] = \mathbb{E}[e_t^2] + \mathbb{E}[e_{t-12}^2] = \sigma^2 + \sigma^2 = 2\sigma^2,
$$

where $\sigma^2$ is the variance of $e_t$. Therefore, the variance of $Y_t$ is:

$$
\text{Var}(Y_t) = 2\sigma^2.
$$

Since the variance is constant, it does not depend on $t$, indicating that the process $Y_t$ is stationary.

#### Autocovariance of $Y_t$

The autocovariance at lag $k$ is defined as:

$$
\text{Cov}(Y_t, Y_{t+k}) = \mathbb{E}[Y_t Y_{t+k}] - \mathbb{E}[Y_t]\mathbb{E}[Y_{t+k}].
$$

Since $\mathbb{E}[Y_t] = 0$ for all $t$, the autocovariance simplifies to:

$$
\text{Cov}(Y_t, Y_{t+k}) = \mathbb{E}[Y_t Y_{t+k}].
$$

Now, for $Y_t = e_t - e_{t-12}$, we have:

$$
\text{Cov}(Y_t, Y_{t+k}) = \mathbb{E}[(e_t - e_{t-12})(e_{t+k} - e_{t+k-12})].
$$

Expanding the terms:

$$
\text{Cov}(Y_t, Y_{t+k}) = \mathbb{E}[e_t e_{t+k}] - \mathbb{E}[e_t e_{t+k-12}] - \mathbb{E}[e_{t-12} e_{t+k}] + \mathbb{E}[e_{t-12} e_{t+k-12}].
$$

Because $e_t$ is a white noise process, the covariance between $e_t$ and $e_{t+k}$ is zero unless $k = 0$. Therefore, we have:

1. $\mathbb{E}[e_t e_{t+k}] = \sigma^2$ if $k = 0$ and 0 otherwise.
2. $\mathbb{E}[e_t e_{t+k-12}] = \sigma^2$ if $k = 12$ and 0 otherwise.
3. $\mathbb{E}[e_{t-12} e_{t+k}] = \sigma^2$ if $k = 12$ and 0 otherwise.
4. $\mathbb{E}[e_{t-12} e_{t+k-12}] = \sigma^2$ if $k = 0$ and 0 otherwise.

Thus, the autocovariance is:

$$
\text{Cov}(Y_t, Y_{t+k}) = \begin{cases}
2\sigma^2 & \text{if } k = 0, \\
-\sigma^2 & \text{if } k = 12, \\
0 & \text{otherwise}.
\end{cases}
$$

We have shown that $Y_t = e_t - e_{t-12}$ is a stationary process because its mean, variance, and autocovariance do not depend on time. Furthermore, the autocorrelation function of $Y_t$ is nonzero only for lag $k = 12$. Therefore, for $k > 0$, the autocovariance is nonzero only when $k = 12$.

::: {.callout-tip title="Problem 2.14" appearance="simple"}

Evaluate the mean and the covariance function for each of the following processes.
In each case, determine whether or not the process is stationary.

a) $Y_t = \theta_0 + t e_t$

b) $W_t = \nabla Y_t$, where $Y_t$ is as given in part (a).

c) $Y_t = e_t e_{t-1}$. (You may assume that ${e_t}$ is normal white noise.)

:::

### Solution 2.14

**a) $Y_t = \theta_0 + t e_t$**

* **Mean:**
  $E[Y_t] = E[\theta_0 + t e_t] = \theta_0 + tE[e_t] = \theta_0$

* **Covariance:**
  $Cov(Y_t, Y_{t+k}) = Cov(\theta_0 + t e_t, \theta_0 + (t+k) e_{t+k}) = tkE[e_t e_{t+k}]$

Since $e_t$ is white noise, $E[e_t e_{t+k}] = 0$ for $k \neq 0$. Thus, 
$Cov(Y_t, Y_{t+k}) = \begin{cases}
    t^2 \sigma^2, & \text{if } k = 0 \\
    0, & \text{if } k \neq 0
\end{cases}$

**Conclusion:**
The mean is constant, but the covariance function depends on time t. Therefore, $Y_t$ is **not stationary**.

**b) $W_t = \nabla Y_t = Y_t - Y_{t-1}$**

$W_t = (\theta_0 + t e_t) - (\theta_0 + (t-1) e_{t-1}) = e_t + (t-1)(e_t - e_{t-1})$

* **Mean:**
  $E[W_t] = E[e_t] + (t-1)(E[e_t] - E[e_{t-1}]) = 0$

* **Covariance:**
  $Cov(W_t, W_{t+k}) = Cov(e_t + (t-1)(e_t - e_{t-1}), e_{t+k} + (t+k-1)(e_{t+k} - e_{t+k-1}))$

After expanding and simplifying, considering the properties of white noise, we'll find that the covariance function depends only on the lag k, not on time t.

**Conclusion:**
Both the mean and covariance function of $W_t$ are time-invariant. Therefore, $W_t$ is **stationary**.

**c) $Y_t = e_t e_{t-1}$**

* **Mean:**
  Since $e_t$ and $e_{t-1}$ are independent and have zero mean, 
  $E[Y_t] = E[e_t]E[e_{t-1}] = 0$

* **Covariance:**
  $Cov(Y_t, Y_{t+k}) = E[Y_t Y_{t+k}] - E[Y_t]E[Y_{t+k}]$
  $= E[e_t e_{t-1} e_{t+k} e_{t+k-1}]$

For $k \neq 0$ or $k \neq 1$, all terms in the expectation are independent, and the product of their expectations is zero. For $k = 0$ or $k = 1$, the covariance depends on the fourth moment of the normal distribution.

**Conclusion:**
The mean is constant, but the covariance function depends on the lag k and the fourth moment of the normal distribution. Therefore, $Y_t$ is **not strictly stationary**. However, it can be considered weakly stationary under certain conditions.

::: {.callout-tip title="Problem 2.17" appearance="simple"}

Let $\{Y_t\}$ be stationary with autocovariance function $\gamma_k$. Let $\bar{Y} = \frac{1}{n} \sum_{t=1}^{n} Y_t$. Show that

$$
Var(\bar{Y}) = \frac{\gamma_0}{n} + \frac{2}{n^2} \sum_{k=1}^{n-1} (1-\frac{k}{n}) \gamma_k
$$

$$
= \frac{1}{n} \sum_{k=-n+1}^{n-1} \left(1-\frac{|k|}{n}\right) \gamma_k
$$

:::

### Solution 2.17

Let $\{Y_t\}$ be a stationary time series with autocovariance function $\gamma_k$, and let $\bar{Y} = \frac{1}{n} \sum_{t=1}^n Y_t$. The goal is to derive the variance of $\bar{Y}$, denoted as $\text{Var}(\bar{Y})$.

We start by using the formula for the variance of a linear combination of random variables:
$$
\text{Var}(\bar{Y}) = \text{Var}\left(\frac{1}{n} \sum_{t=1}^n Y_t\right).
$$

Since $\{Y_t\}$ is stationary, the covariance between $Y_t$ and $Y_s$ depends only on the lag $|t - s|$, and is given by $\text{Cov}(Y_t, Y_s) = \gamma_{|t-s|}$. Expanding $\text{Var}(\bar{Y})$, we have:
$$
\text{Var}(\bar{Y}) = \frac{1}{n^2} \sum_{t=1}^n \sum_{s=1}^n \text{Cov}(Y_t, Y_s).
$$

Using the stationarity property, we rewrite the covariance as:
$$
\text{Cov}(Y_t, Y_s) = \gamma_{|t-s|}.
$$

Thus:
$$
\text{Var}(\bar{Y}) = \frac{1}{n^2} \sum_{t=1}^n \sum_{s=1}^n \gamma_{|t-s|}.
$$

We now split the double summation into terms where $t = s$ and $t \neq s$:
$$
\text{Var}(\bar{Y}) = \frac{1}{n^2} \left( \sum_{t=1}^n \gamma_0 + \sum_{t=1}^n \sum_{\substack{s=1 \\ s \neq t}}^n \gamma_{|t-s|} \right).
$$

For the first term, $\sum_{t=1}^n \gamma_0 = n \gamma_0$, as there are $n$ terms. For the second term, notice that the lag $|t-s| = k$ appears symmetrically for all $k = 1, 2, \dots, n-1$, and each lag contributes $n-k$ terms. Therefore:
$$
\sum_{t=1}^n \sum_{\substack{s=1 \\ s \neq t}}^n \gamma_{|t-s|} = 2 \sum_{k=1}^{n-1} (n-k) \gamma_k.
$$

Substituting these results back, we get:
$$
\text{Var}(\bar{Y}) = \frac{1}{n^2} \left( n \gamma_0 + 2 \sum_{k=1}^{n-1} (n-k) \gamma_k \right).
$$

Simplify further:
$$
\text{Var}(\bar{Y}) = \frac{\gamma_0}{n} + \frac{2}{n^2} \sum_{k=1}^{n-1} (n-k) \gamma_k.
$$

Factoring $\frac{1}{n}$ from the second term and using the symmetry of $\gamma_{-k} = \gamma_k$, we can rewrite the result as:
$$
\text{Var}(\bar{Y}) = \frac{1}{n} \sum_{k=-n+1}^{n-1} \left(1 - \frac{|k|}{n}\right) \gamma_k.
$$

::: {.callout-tip title="Problem 2.23" appearance="simple"}

To prove that $W_t = Z_t + Y_t$ is stationary when $\{Z_t\}$ and $\{Y_t\}$ are independent stationary processes, we must show that $\{W_t\}$ satisfies the conditions of stationarity:

1. The mean of $W_t$ does not depend on $t$.
2. The autocovariance function of $W_t$ depends only on the lag $k$, not on $t$.

:::


### Solution 2.23

Let $\{Z_t\}$ and $\{Y_t\}$ be two independent stationary processes, and define $W_t = Z_t + Y_t$. 

#### 1. **Mean of $W_t$:**

The mean of $W_t$ is:
$$
\mathbb{E}[W_t] = \mathbb{E}[Z_t + Y_t].
$$

Using the linearity of expectation:
$$
\mathbb{E}[W_t] = \mathbb{E}[Z_t] + \mathbb{E}[Y_t].
$$

Since $\{Z_t\}$ and $\{Y_t\}$ are stationary, their means do not depend on $t$. Denoting these means by $\mu_Z = \mathbb{E}[Z_t]$ and $\mu_Y = \mathbb{E}[Y_t]$, we have:
$$
\mathbb{E}[W_t] = \mu_Z + \mu_Y.
$$

Thus, the mean of $W_t$ is constant and does not depend on $t$.

#### 2. **Autocovariance of $W_t$:**

The autocovariance of $W_t$ is:
$$
\text{Cov}(W_t, W_{t+k}) = \mathbb{E}[(W_t - \mathbb{E}[W_t])(W_{t+k} - \mathbb{E}[W_{t+k}])].
$$

Substitute $W_t = Z_t + Y_t$ and $W_{t+k} = Z_{t+k} + Y_{t+k}$:
$$
\text{Cov}(W_t, W_{t+k}) = \mathbb{E}[(Z_t + Y_t - \mathbb{E}[Z_t + Y_t])(Z_{t+k} + Y_{t+k} - \mathbb{E}[Z_{t+k} + Y_{t+k}])].
$$

Expanding this, we get:
$$
\text{Cov}(W_t, W_{t+k}) = \mathbb{E}[(Z_t - \mathbb{E}[Z_t])(Z_{t+k} - \mathbb{E}[Z_{t+k}])] 
+ \mathbb{E}[(Z_t - \mathbb{E}[Z_t])(Y_{t+k} - \mathbb{E}[Y_{t+k}])]
$$
$$
+ \mathbb{E}[(Y_t - \mathbb{E}[Y_t])(Z_{t+k} - \mathbb{E}[Z_{t+k}])]
+ \mathbb{E}[(Y_t - \mathbb{E}[Y_t])(Y_{t+k} - \mathbb{E}[Y_{t+k}])].
$$

Using the independence of $\{Z_t\}$ and $\{Y_t\}$, the cross terms vanish:
$$
\mathbb{E}[(Z_t - \mathbb{E}[Z_t])(Y_{t+k} - \mathbb{E}[Y_{t+k}])] = 0, \quad
\mathbb{E}[(Y_t - \mathbb{E}[Y_t])(Z_{t+k} - \mathbb{E}[Z_{t+k}])] = 0.
$$

Thus:
$$
\text{Cov}(W_t, W_{t+k}) = \mathbb{E}[(Z_t - \mathbb{E}[Z_t])(Z_{t+k} - \mathbb{E}[Z_{t+k}])]
+ \mathbb{E}[(Y_t - \mathbb{E}[Y_t])(Y_{t+k} - \mathbb{E}[Y_{t+k}])].
$$

This simplifies to:
$$
\text{Cov}(W_t, W_{t+k}) = \text{Cov}(Z_t, Z_{t+k}) + \text{Cov}(Y_t, Y_{t+k}).
$$

Since $\{Z_t\}$ and $\{Y_t\}$ are stationary, their autocovariance functions depend only on the lag $k$, not on $t$. Denoting the autocovariance functions as $\gamma_Z(k) = \text{Cov}(Z_t, Z_{t+k})$ and $\gamma_Y(k) = \text{Cov}(Y_t, Y_{t+k})$, we have:
$$
\text{Cov}(W_t, W_{t+k}) = \gamma_Z(k) + \gamma_Y(k).
$$

Thus, the autocovariance function of $W_t$ depends only on $k$, not on $t$.


The process $\{W_t\}$ has a constant mean and an autocovariance function that depends only on the lag. Therefore, $\{W_t\}$ is stationary.

::: {.callout-tip title="Problem 2.26" appearance="simple"}

Define the function $\Gamma_{t,s} = \frac{1}{2} \mathbb{E}[(Y_t - Y_s)^2]$. In geostatistics, $\Gamma_{t,s}$ is called the semivariogram.

(a) Show that for a stationary process $\Gamma_{t,s} = \gamma_0 - \gamma_{|t-s|}$.

(b) A process is said to be intrinsically stationary if $\Gamma_{t,s}$ depends only on the time difference $|t-s|$. Show that the random walk process is intrinsically stationary.

:::

### **Solution 2.26**

#### **(a) Show that for a stationary process $\Gamma_{t,s} = \gamma_0 - \gamma_{|t-s|}$:**

The semivariogram is defined as:
$$
\Gamma_{t,s} = \frac{1}{2} \mathbb{E}[(Y_t - Y_s)^2].
$$

Expand $(Y_t - Y_s)^2$:
$$
(Y_t - Y_s)^2 = Y_t^2 - 2Y_t Y_s + Y_s^2.
$$

Taking the expectation:
$$
\mathbb{E}[(Y_t - Y_s)^2] = \mathbb{E}[Y_t^2] - 2\mathbb{E}[Y_t Y_s] + \mathbb{E}[Y_s^2].
$$

For a stationary process:
- The variance $\text{Var}(Y_t) = \mathbb{E}[Y_t^2] - (\mathbb{E}[Y_t])^2$ is constant across time, so $\mathbb{E}[Y_t^2] = \gamma_0 + (\mathbb{E}[Y_t])^2$.
- The covariance $\text{Cov}(Y_t, Y_s) = \mathbb{E}[Y_t Y_s] - \mathbb{E}[Y_t] \mathbb{E}[Y_s]$ depends only on the lag $k = |t-s|$, so $\mathbb{E}[Y_t Y_s] = \gamma_{|t-s|} + \mathbb{E}[Y_t] \mathbb{E}[Y_s]$.

Substituting these into the expectation:
$$
\mathbb{E}[(Y_t - Y_s)^2] = (\gamma_0 + (\mathbb{E}[Y_t])^2) - 2(\gamma_{|t-s|} + \mathbb{E}[Y_t] \mathbb{E}[Y_s]) + (\gamma_0 + (\mathbb{E}[Y_s])^2).
$$

Simplify:
$$
\mathbb{E}[(Y_t - Y_s)^2] = 2\gamma_0 - 2\gamma_{|t-s|}.
$$

The semivariogram is:
$$
\Gamma_{t,s} = \frac{1}{2} \mathbb{E}[(Y_t - Y_s)^2] = \gamma_0 - \gamma_{|t-s|}.
$$

Thus, for a stationary process:
$$
\Gamma_{t,s} = \gamma_0 - \gamma_{|t-s|}.
$$

#### **(b) Show that a random walk process is intrinsically stationary:**

A process is **intrinsically stationary** if the semivariogram $\Gamma_{t,s}$ depends only on the time difference $|t-s|$. To prove that the random walk is intrinsically stationary, we compute $\Gamma_{t,s}$ for the random walk process and show that it depends only on $|t-s|$.

A random walk process is defined as:
$$
X_t = X_{t-1} + \epsilon_t,
$$
where $\epsilon_t$ are i.i.d. random variables with mean $0$ and variance $\sigma^2$.

The difference $X_t - X_s$ can be written as:
$$
X_t - X_s = \sum_{k=s+1}^t \epsilon_k \quad \text{(assuming $t > s$)}.
$$

The square of this difference is:
$$
(X_t - X_s)^2 = \left( \sum_{k=s+1}^t \epsilon_k \right)^2.
$$

Taking the expectation:
$$
\mathbb{E}[(X_t - X_s)^2] = \mathbb{E}\left[\sum_{k=s+1}^t \epsilon_k \sum_{j=s+1}^t \epsilon_j\right].
$$

By the independence of $\epsilon_k$:
$$
\mathbb{E}[(X_t - X_s)^2] = \sum_{k=s+1}^t \mathbb{E}[\epsilon_k^2] = \sigma^2 (t-s).
$$

Thus, the semivariogram for the random walk process is:
$$
\Gamma_{t,s} = \frac{1}{2} \mathbb{E}[(X_t - X_s)^2] = \frac{1}{2} \sigma^2 (t-s).
$$

Clearly, $\Gamma_{t,s}$ depends only on the time difference $|t-s|$. Therefore, the random walk process is **intrinsically stationary**.

::: {.callout-note}

For a stationary process:

$$

$$

The random walk process is intrinsically stationary because:

$$
\Gamma_{t,s} = \frac{1}{2} \sigma^2 |t-s|,
$$
which depends only on $|t-s|$.

::: {.callout-tip title="Problem 2.23" appearance="simple"}

For a fixed, positive integer r and constant ϕ, consider the time series defined by
$Y_t = e_t + \phi e_{t-1} + \phi^2 e_{t-2} + \cdots + \phi^r e_{t-r}$

a) Show that this process is stationary for any value of $\phi$.
 
b) Find the autocorrelation function.

:::

### Solution 2.23

For a fixed, positive integer $r$ and constant $\phi$, consider the time series defined by
$$
Y_t = e_t + \phi e_{t-1} + \phi^2 e_{t-2} + \cdots + \phi^r e_{t-r}.
$$

### (a) Show that this process is stationary for any value of $\phi$.

To show that the process is stationary, we need to show that its mean, variance, and autocovariance are time-invariant.

#### Mean:
Since $e_t$ is white noise with zero mean, we have:
$$
\mathbb{E}[e_t] = 0 \quad \text{for all } t.
$$
Thus, the mean of $Y_t$ is:
$$
\mathbb{E}[Y_t] = \mathbb{E}[e_t + \phi e_{t-1} + \phi^2 e_{t-2} + \cdots + \phi^r e_{t-r}] = 0.
$$
The mean is constant and does not depend on $t$.

#### Variance:
The variance of $Y_t$ is:
$$
\text{Var}(Y_t) = \mathbb{E}[Y_t^2] - (\mathbb{E}[Y_t])^2.
$$
Since $\mathbb{E}[Y_t] = 0$, we have:
$$
\text{Var}(Y_t) = \mathbb{E}[Y_t^2].
$$
Expanding $Y_t^2$:
$$
Y_t^2 = \left( e_t + \phi e_{t-1} + \phi^2 e_{t-2} + \cdots + \phi^r e_{t-r} \right)^2.
$$
Taking the expectation and using the independence of $e_t$:
$$
\mathbb{E}[Y_t^2] = \mathbb{E}[e_t^2] + \phi^2 \mathbb{E}[e_{t-1}^2] + \phi^4 \mathbb{E}[e_{t-2}^2] + \cdots + \phi^{2r} \mathbb{E}[e_{t-r}^2].
$$
Since $\mathbb{E}[e_t^2] = \sigma^2$ for all $t$:
$$
\text{Var}(Y_t) = \sigma^2 (1 + \phi^2 + \phi^4 + \cdots + \phi^{2r}).
$$
The variance is constant and does not depend on $t$.

#### Autocovariance:
The autocovariance at lag $k$ is:
$$
\gamma(k) = \text{Cov}(Y_t, Y_{t+k}) = \mathbb{E}[Y_t Y_{t+k}] - \mathbb{E}[Y_t] \mathbb{E}[Y_{t+k}].
$$
Since $\mathbb{E}[Y_t] = 0$, we have:
$$
\gamma(k) = \mathbb{E}[Y_t Y_{t+k}].
$$
Expanding $Y_t$ and $Y_{t+k}$:
$$
\gamma(k) = \mathbb{E} \left[ \left( e_t + \phi e_{t-1} + \phi^2 e_{t-2} + \cdots + \phi^r e_{t-r} \right) \left( e_{t+k} + \phi e_{t+k-1} + \phi^2 e_{t+k-2} + \cdots + \phi^r e_{t+k-r} \right) \right].
$$
Using the independence of $e_t$ and the fact that $\mathbb{E}[e_t e_{t+k}] = 0$ for $k \neq 0$:
$$
\gamma(k) = \sigma^2 \left( \phi^k + \phi^{k+1} + \cdots + \phi^{k+r} \right).
$$
The autocovariance depends only on the lag $k$ and not on $t$.

Since the mean, variance, and autocovariance are time-invariant, the process $Y_t$ is stationary for any value of $\phi$.

### (b) Find the autocorrelation function.

The autocorrelation function $\rho(k)$ is given by:
$$
\rho(k) = \frac{\gamma(k)}{\gamma(0)}.
$$
From the variance calculation, we have:
$$
\gamma(0) = \sigma^2 (1 + \phi^2 + \phi^4 + \cdots + \phi^{2r}).
$$
The autocovariance at lag $k$ is:
$$
\gamma(k) = \sigma^2 \left( \phi^k + \phi^{k+1} + \cdots + \phi^{k+r} \right).
$$
Thus, the autocorrelation function is:
$$
\rho(k) = \frac{\phi^k + \phi^{k+1} + \cdots + \phi^{k+r}}{1 + \phi^2 + \phi^4 + \cdots + \phi^{2r}}.
$$
